# Conditional Statements

**If Statement** 


If condition is true then execute the statement within **if** block

```python
# An "if statement" is written by using the if keyword. 
a = 33
b = 200
if (b > a):
  print("b is greater than a")
```

** Elif Statement **

```python
# The elif keyword is pythons way of saying "if the previous conditions were not true, then try this condition"

a = 45
b = 45
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
```

** Else Statement**

```python
#  The else keyword catches anything which isn't caught by the preceding conditions.

a = 100
b = 23
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

```python
# we can also have an else without the elif
a = 200
b = 33
if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```

# Python Loops

Python has two primitive loop commands:

#while loops
#for loops

# The while Loop

With the while loop we can execute a set of statements as long as a condition is true.

#In Python, the body of the while loop is determined through indentation.

#Body starts with indentation and the first unindented line marks the end.

#Python interprets any non-zero value as True. None and 0 are interpreted as False.




```python
# Print i as long as i is less than 6:
#Note: remember to increment i, or else the loop will continue forever.

i = 1
while i < 6:
  print(i)
  i += 1
```

**For-Loop**

The for..in statement is another looping statement which iterates over a sequence of objects i.e. go through each item in a sequence. A sequence is just an ordered collection of items.



```python
for i in range(1,5,2):
    print (i)
    break
else:
   print ("The for loop is over")
```

```python
# Traversing a list   +++++++++++++++++++
# Guess What is the output
numbers = [2, 3, 5]
sum = 0
for num in numbers:
    sum += num
print (sum)
```

**Break statement**

The break statement is used to break out of a loop statement i.e. stop the execution of a looping statement, even if the loop condition has not become False or the sequence of items has not been completely iterated over.

**Note:** if you break out of a for or while loop, any corresponding loop else block is not executed.

```python
for i in range(1,10):
       if i == 5:
           break
       print (i)
print('Done')
```

**Continue Statement**

Continue statement will continue the execution even if condition is met

```python
i=0
while i < 10 :
 i+=1
 if i==5 :
  continue
 print(i)
```
